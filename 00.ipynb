### Random tensors

```python
# Create a random tensor of size (3,4)
random_tensor = torch.rand(3,4)
random_tensor
```

```python
random_tensor.ndim
```

```python
# Create a random tensor with similiar shape to an image tensor
random_image_size_tensor = torch.rand(size=(3, 224, 224)) #height, width, colour channel
random_image_size_tensor.shape, random_image_size_tensor.ndim
```

### Zeros and ones

```python
# Create a tensor of all zeros
zeros = torch.zeros(size=(3,4))
zeros
```

```python
zeros*random_tensor
```

```python
# Create a tensor of all ones
ones = torch.ones(size=(3,4))
ones
```

```python
ones.dtype
```

### Create a range of tensors and tensors-like

```python
# Use torch.arange()
one_to_ten = torch.arange(start=0,end=10, step=1)
one_to_ten
```

```python
# Creating tensors like
ten_zeros = torch.zeros_like(input=random_tensor)
ten_zeros
```

### Tensor dType

```python
float_32_tensor = torch.tensor([3.0, 6.0, 9.0],
                               dtype=None,
                               device=None,
                               requires_grad=True)
float_32_tensor
```
